# Neural Network Activation Functions and Visualizations

- Sigmoid

- Hyperbolic tangent

- ReLU

- Leaky ReLU

- ELU

- Swish

- Softmax

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
x = np.linspace(start = -10, stop = 10, num = 10000, endpoint = True)
````

### Sigmoid activation function

$f\left(x\right)=\frac{1}{1+e^{-x}}$

````python
x_sigmoid = 1 / (1 + np.exp(-x))
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_sigmoid)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', c = 'black')
plt.axvline(x = 0, ls = '--', c = 'black')
plt.title("Sigmoid activation function")
plt.show()
````

### Hyperbolic Tangent activation function


$f\left(x\right)=\frac{e^x - e^{-x}}{e^x + e^{-x}}$

````python
x_tanh = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_tanh)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', c = 'black')
plt.axvline(x = 0, ls = '--', c = 'black')
plt.title("Tanh activation function")
plt.show()
````

### Rectified Linear Unit (ReLU) activation function

$ReLU(x) = max(0, x)$

```
lambda <arguments> : <Return Value if condition is True> if <condition> else <Return Value if condition is False>
```

For example let’s create a lambda function to check if given value is between 10 to 20-
```
lambda x : True if (x > 10 and x < 20) else False
```

Compute ReLU as-
```
relu = lambda x: x if x > 0 else 0
```

````python
# numpy implementation instead of lambda function-
x_relu = np.where(x > 0, x, 0)
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_relu, lw = 5, c = 'blue')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', c = 'black')
plt.axvline(x = 0, ls = '--', c = 'black')
plt.title("ReLU activation function")
plt.show()
````

### Leaky ReLU activation function

If $x > 0$, then $x$, else, $\alpha \times x$

````python
alpha = 0.01
````

````python
x_leaky_relu = np.where(x <= 0, (alpha * x), x) 
# x_leaky_relu = np.where(x > 0, x, (alpha * x))
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_leaky_relu, lw = 1, c = 'blue')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', lw = 0.5, c = 'black')
plt.axvline(x = 0, ls = '--', lw = 0.5, c = 'black')
plt.title("Leaky ReLU activation function")
plt.show()
````

### Exponential Linear Unit (ELU) activation function

If $x > 0$, then $x$, else, $\alpha \times (e^x - 1)$

````python
alpha = 1
````

````python
x_elu = np.where(x > 0, x, (alpha * (np.exp(x) - 1)))
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_elu, lw = 1, c = 'blue')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', lw = 0.5, c = 'black')
plt.axvline(x = 0, ls = '--', lw = 0.5, c = 'black')
plt.title("Exponential Linear Unit activation function")
plt.show()
````

### Swish activation function

$f(x) = \frac{x}{1 + e^{-\beta \times x}}$ = input x sigmoid(beta x input)

````python
x = np.linspace(start = -10, stop = 10, num = 10000, endpoint = True)
````

````python
beta = 1
````

````python
x_swish = x / (1 + np.exp(beta * -x))
````

````python
plt.figure(figsize = (8, 7))
plt.plot(x, x_swish, lw = 1, c = 'blue')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0, ls = '--', lw = 0.5, c = 'black')
plt.axvline(x = 0, ls = '--', lw = 0.5, c = 'black')
plt.title("Swish activation function (beta = 1)")
plt.show()
````

### Softmax activation function

````python
# Predictions/raw scores/logits-
pred = np.random.randn(10)
````

````python
softmax_act = np.exp(pred) / np.sum(np.exp(pred))
````

````python
np.sort(pred)
````

````python
np.sort(softmax_act)
````

````python
plt.figure(figsize = (8, 7))
# plt.bar(pred, softmax_act)
sns.barplot(x = pred, y = softmax_act)
plt.xlabel("raw scores/logits")
plt.ylabel("softmax probabilities")
plt.title("Softmax activation function")
plt.xticks(rotation = 90)
plt.show()
````
